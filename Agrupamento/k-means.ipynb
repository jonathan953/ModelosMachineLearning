# üå∏ Clusteriza√ß√£o com K-Means no Dataset Iris

- Pr√©-processamento com normaliza√ß√£o
- Avalia√ß√£o do n√∫mero ideal de clusters com **5 m√©tricas diferentes**
- Visualiza√ß√£o interativa dos resultados usando **gr√°ficos explicativos** e **PCA**

## üìÅ Estrutura do Projeto

O projeto segue o fluxo padr√£o de Ci√™ncia de Dados:

### 1. üì• Carregamento dos Dados

- Gera√ß√£o de heatmap de correla√ß√£o entre vari√°veis

---

### 4. üìâ Defini√ß√£o do N√∫mero Ideal de Clusters (K)

O projeto implementa **cinco abordagens quantitativas** para ajudar na escolha do n√∫mero ideal de clusters:

#### üî∏ Elbow Method (M√©todo do Cotovelo)
- Baseado no WCSS (Within-Cluster Sum of Squares)
- O ponto de inflex√£o na curva representa o K √≥timo
- Gr√°fico explicativo inclu√≠do com `print()` logo abaixo

#### üî∏ Silhouette Score
- Mede qu√£o bem os pontos se ajustam ao seu cluster comparado com os outros
- Quanto mais pr√≥ximo de 1, melhor
- Gr√°fico e explica√ß√£o inclu√≠dos no c√≥digo

#### üî∏ Calinski-Harabasz Index
- Avalia a raz√£o entre separa√ß√£o inter-cluster e coes√£o intra-cluster
- Quanto maior, melhor

#### üî∏ Davies-Bouldin Index
- Mede a sobreposi√ß√£o entre clusters
- Quanto menor, melhor

#### üî∏ Dist√¢ncias Intra e Inter-Cluster
- **Intra-cluster**: mede a coes√£o interna (quanto menor, melhor)
- **Inter-cluster**: mede a separa√ß√£o entre clusters (quanto maior, melhor)

Todos os gr√°ficos s√£o exibidos separadamente com **coment√°rios descritivos abaixo de cada um**, explicando como interpretar as m√©tricas.

---

### 5. üß† Treinamento do Modelo

- Aplica `KMeans(n_clusters=best_k)` com o valor de K considerado ideal
- Os r√≥tulos de cluster s√£o adicionados √† base como uma nova coluna `cluster`

---

### 6. üîÑ Redu√ß√£o de Dimensionalidade com PCA

### 7. üé® Visualiza√ß√£o Final

## üìä An√°lise dos Resultados

O script imprime uma tabela com os resultados de todas as m√©tricas para K entre 2 e 10, seguida de uma explica√ß√£o:

```text
üìä Como avaliar os resultados por K:

Cada linha da tabela representa um n√∫mero de clusters (K) e os valores de desempenho para 5 m√©tricas:

‚û°Ô∏è Silhouette Score: mede o qu√£o bem os pontos est√£o agrupados. Quanto MAIOR, melhor.
‚û°Ô∏è Calinski-Harabasz Index: mede a separa√ß√£o entre os clusters. Quanto MAIOR, melhor.
‚û°Ô∏è Davies-Bouldin Index: mede a similaridade entre clusters. Quanto MENOR, melhor.
‚û°Ô∏è Intra-cluster Distance: mede a coes√£o interna. Quanto MENOR, melhor.
‚û°Ô∏è Inter-cluster Distance: mede a separa√ß√£o entre clusters. Quanto MAIOR, melhor.

‚úÖ K = 2 se destaca como o melhor em praticamente todas as m√©tricas:
   - Maior Silhouette (0.58)
   - Maior Calinski (251)
   - Menor Davies-Bouldin (0.59)
   - Maior separa√ß√£o entre clusters (3.36)

‚ö†Ô∏è Por√©m, o dataset Iris tem 3 esp√©cies reais. Se voc√™ quer refletir essa estrutura real, K = 3 tamb√©m √© uma boa escolha, mesmo que tenha desempenho um pouco inferior nas m√©tricas.

üéØ Em resumo:
- Use K = 2 para m√°xima performance matem√°tica.
- Use K = 3 para refletir a separa√ß√£o real das esp√©cies do Iris.


````python
# üîÅ Projeto completo: Iris + KMeans + PCA + M√∫ltiplas M√©tricas + Explica√ß√µes Gr√°ficas

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import (
    silhouette_score,
    calinski_harabasz_score,
    davies_bouldin_score
)
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
os.environ["LOKY_MAX_CPU_COUNT"] = "4"

# üì• Carregar o dataset Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# üîç An√°lise explorat√≥ria
print("üìå .info():")
print(df.info())
print("\nüìå Missing values:")
print(df.isnull().sum())
print("\nüìå Estat√≠sticas descritivas:")
print(df.describe())
print("\nüìå Correla√ß√£o:")
print(df.corr())
print("\nüìå Primeiras linhas:")
print(df.head())

# üìà Visualiza√ß√µes iniciais
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='viridis')
plt.title("Correla√ß√£o entre vari√°veis - Iris")
plt.show()

sns.pairplot(df)
plt.suptitle("Dispers√µes e distribui√ß√µes - Dataset Iris", y=1.02)
plt.show()

# ‚öôÔ∏è Normaliza√ß√£o
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# üìâ M√©todo do Cotovelo (WCSS)
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("N√∫mero de Clusters (K)")
plt.ylabel("WCSS")
plt.title("M√©todo do Cotovelo - Iris")
plt.grid(True)
plt.show()

print("""
üìâ M√©todo do Cotovelo (Elbow Method):
Esse gr√°fico mostra a soma das dist√¢ncias quadradas dentro dos clusters (WCSS) para diferentes valores de K.

‚úÖ O ponto ideal √© onde ocorre uma "curva" ou "cotovelo" ‚Äî ou seja, onde o ganho de adicionar novos clusters passa a ser marginal.
‚ùå Ap√≥s esse ponto, continuar aumentando o n√∫mero de clusters reduz muito pouco o WCSS e pode gerar overfitting.

üîç Use o gr√°fico para visualizar onde o WCSS come√ßa a diminuir mais lentamente.
""")

# üìä Avalia√ß√£o de m√∫ltiplas m√©tricas
k_values = range(2, 11)
silhouette_scores = []
calinski_scores = []
davies_scores = []
intra_cluster_distances = []
inter_cluster_distances = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(scaled_data)

    silhouette_scores.append(silhouette_score(scaled_data, labels))
    calinski_scores.append(calinski_harabasz_score(scaled_data, labels))
    davies_scores.append(davies_bouldin_score(scaled_data, labels))

    intra_dist = np.mean([
        np.linalg.norm(scaled_data[i] - kmeans.cluster_centers_[label])
        for i, label in enumerate(labels)
    ])
    intra_cluster_distances.append(intra_dist)

    centers = kmeans.cluster_centers_
    inter_distances = [
        np.linalg.norm(a - b)
        for i, a in enumerate(centers)
        for j, b in enumerate(centers) if i < j
    ]
    inter_cluster_distances.append(np.mean(inter_distances))

# üéØ Gr√°fico + explica√ß√£o: Silhouette Score
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores, marker='o', color='green')
plt.title("Silhouette Score")
plt.xlabel("N√∫mero de Clusters (K)")
plt.ylabel("Score")
plt.grid(True)
plt.show()
print("""
üü¢ Silhouette Score:
Esse gr√°fico mostra o qu√£o bem os pontos est√£o agrupados.

‚úÖ Quanto mais pr√≥ximo de 1, melhor o agrupamento.
‚ùå Valores abaixo de 0.5 indicam que os clusters podem estar se sobrepondo.
üîç Escolha o valor de K com o maior Silhouette Score.
""")

# üéØ Gr√°fico + explica√ß√£o: Calinski-Harabasz
plt.figure(figsize=(8, 5))
plt.plot(k_values, calinski_scores, marker='o', color='blue')
plt.title("Calinski-Harabasz Index")
plt.xlabel("N√∫mero de Clusters (K)")
plt.ylabel("Score")
plt.grid(True)
plt.show()
print("""
üîµ Calinski-Harabasz Index:
Avalia a separa√ß√£o entre clusters e a compacidade dentro de cada grupo.

‚úÖ Quanto maior o valor, melhor os clusters est√£o separados e definidos.
üìå √ìtimo para refor√ßar a escolha de K que aparece em outras m√©tricas.
""")

# üéØ Gr√°fico + explica√ß√£o: Davies-Bouldin
plt.figure(figsize=(8, 5))
plt.plot(k_values, davies_scores, marker='o', color='red')
plt.title("Davies-Bouldin Index")
plt.xlabel("N√∫mero de Clusters (K)")
plt.ylabel("√çndice")
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
print("""
üî¥ Davies-Bouldin Index:
Mede a similaridade entre os clusters.

‚úÖ Quanto menor o √≠ndice, melhor.
‚ùå Valores altos significam que os clusters est√£o muito pr√≥ximos e mal definidos.
üîç Escolha o menor valor poss√≠vel para definir o melhor K.
""")

# üéØ Gr√°fico + explica√ß√£o: Dist√¢ncia Intra-Cluster
plt.figure(figsize=(8, 5))
plt.plot(k_values, intra_cluster_distances, marker='o', color='purple')
plt.title("Dist√¢ncia Intra-Cluster M√©dia")
plt.xlabel("N√∫mero de Clusters (K)")
plt.ylabel("Dist√¢ncia M√©dia")
plt.grid(True)
plt.show()
print("""
üü£ Dist√¢ncia Intra-Cluster:
Mede o qu√£o pr√≥ximos os pontos est√£o do centro de seu cluster.

‚úÖ Valores baixos indicam que os pontos est√£o bem agrupados (alta coes√£o).
‚ùå Redu√ß√µes muito pequenas ao aumentar K indicam que mais clusters n√£o est√£o ajudando.
""")

# üéØ Gr√°fico + explica√ß√£o: Dist√¢ncia Inter-Cluster
plt.figure(figsize=(8, 5))
plt.plot(k_values, inter_cluster_distances, marker='o', color='orange')
plt.title("Dist√¢ncia Inter-Cluster M√©dia")
plt.xlabel("N√∫mero de Clusters (K)")
plt.ylabel("Dist√¢ncia M√©dia")
plt.grid(True)
plt.show()
print("""
üü† Dist√¢ncia Inter-Cluster:
Mede o qu√£o distantes os clusters est√£o entre si.

‚úÖ Quanto maior a dist√¢ncia, mais separados est√£o os grupos.
‚ùå Valores baixos sugerem sobreposi√ß√£o entre clusters.
üîç Boa para avaliar se os clusters s√£o bem distintos.
""")

# üìã Resultados num√©ricos
resultados_df = pd.DataFrame({
    "K": list(k_values),
    "Silhouette": silhouette_scores,
    "Calinski-Harabasz": calinski_scores,
    "Davies-Bouldin": davies_scores,
    "Intra-cluster": intra_cluster_distances,
    "Inter-cluster": inter_cluster_distances
})
print("\nüìä Tabela de Resultados por K:")
print(resultados_df)

print("""
üìä Como avaliar os resultados por K:

Cada linha da tabela representa um n√∫mero de clusters (K) e os valores de desempenho para 5 m√©tricas:

‚û°Ô∏è Silhouette Score: mede o qu√£o bem os pontos est√£o agrupados. Quanto MAIOR, melhor.
‚û°Ô∏è Calinski-Harabasz Index: mede a separa√ß√£o entre os clusters. Quanto MAIOR, melhor.
‚û°Ô∏è Davies-Bouldin Index: mede a similaridade entre clusters. Quanto MENOR, melhor.
‚û°Ô∏è Intra-cluster Distance: mede a coes√£o interna. Quanto MENOR, melhor.
‚û°Ô∏è Inter-cluster Distance: mede a separa√ß√£o entre clusters. Quanto MAIOR, melhor.

‚úÖ K = 2 se destaca como o melhor em praticamente todas as m√©tricas:
   - Maior Silhouette (0.58)
   - Maior Calinski (251)
   - Menor Davies-Bouldin (0.59)
   - Maior dist√¢ncia inter-cluster (3.36)

‚ö†Ô∏è Por√©m, o dataset Iris tem 3 esp√©cies reais. Se voc√™ quer refletir essa estrutura real, K = 3 tamb√©m √© uma boa escolha, mesmo que tenha desempenho um pouco inferior nas m√©tricas.

üéØ Em resumo:
- Use K = 2 se quiser clusters matematicamente mais coesos e bem separados.
- Use K = 3 se quiser refletir a separa√ß√£o real das esp√©cies do Iris.
""")

# ‚úÖ Melhor K com base no Silhouette
best_k = resultados_df["Silhouette"].idxmax() + 2
print(f"\nüî• Melhor K com base no Silhouette Score: {best_k}")

# üß† Aplicar KMeans final
final_kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df['cluster'] = final_kmeans.fit_predict(scaled_data)

# üîÑ PCA para reduzir de 4D para 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
df['pca1'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]

# üé® Visualiza√ß√£o dos clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='pca1', y='pca2', hue='cluster', palette='tab10')
plt.title(f"Clusters com PCA - Iris (K = {best_k})")
plt.show()
````
