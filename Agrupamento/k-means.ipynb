# 🌸 Clusterização com K-Means no Dataset Iris

Este projeto demonstra a aplicação de **clustering não supervisionado com o algoritmo K-Means** no clássico **dataset Iris**. O processo inclui:

- Análise exploratória dos dados (EDA)
- Pré-processamento com normalização
- Avaliação do número ideal de clusters com **5 métricas diferentes**
- Visualização interativa dos resultados usando **gráficos explicativos** e **PCA**

---

## 🔧 Tecnologias Utilizadas

- Python 3.12  
- pandas  
- numpy  
- scikit-learn  
- matplotlib  
- seaborn  

---

## 📁 Estrutura do Projeto

O projeto segue o fluxo padrão de Ciência de Dados:

### 1. 📥 Carregamento dos Dados

Utiliza o `load_iris()` da `sklearn.datasets`, que fornece dados de 150 amostras de flores de 3 espécies com 4 atributos:

- comprimento da sépala  
- largura da sépala  
- comprimento da pétala  
- largura da pétala  

---

### 2. 🔍 Análise Exploratória (EDA)

- Exibição de `.info()`, `.describe()`, verificação de valores nulos e visualização de amostras com `.head()`
- Geração de heatmap de correlação entre variáveis
- Pairplot para observar padrões e possíveis agrupamentos naturais

---

### 3. ⚙️ Pré-processamento

- Normalização dos dados com `StandardScaler()` para remover viés de escala entre atributos, essencial para o algoritmo K-Means

---

### 4. 📉 Definição do Número Ideal de Clusters (K)

O projeto implementa **cinco abordagens quantitativas** para ajudar na escolha do número ideal de clusters:

#### 🔸 Elbow Method (Método do Cotovelo)
- Baseado no WCSS (Within-Cluster Sum of Squares)
- O ponto de inflexão na curva representa o K ótimo
- Gráfico explicativo incluído com `print()` logo abaixo

#### 🔸 Silhouette Score
- Mede quão bem os pontos se ajustam ao seu cluster comparado com os outros
- Quanto mais próximo de 1, melhor
- Gráfico e explicação incluídos no código

#### 🔸 Calinski-Harabasz Index
- Avalia a razão entre separação inter-cluster e coesão intra-cluster
- Quanto maior, melhor

#### 🔸 Davies-Bouldin Index
- Mede a sobreposição entre clusters
- Quanto menor, melhor

#### 🔸 Distâncias Intra e Inter-Cluster
- **Intra-cluster**: mede a coesão interna (quanto menor, melhor)
- **Inter-cluster**: mede a separação entre clusters (quanto maior, melhor)

Todos os gráficos são exibidos separadamente com **comentários descritivos abaixo de cada um**, explicando como interpretar as métricas.

---

### 5. 🧠 Treinamento do Modelo

- Aplica `KMeans(n_clusters=best_k)` com o valor de K considerado ideal
- Os rótulos de cluster são adicionados à base como uma nova coluna `cluster`

---

### 6. 🔄 Redução de Dimensionalidade com PCA

- O PCA reduz os dados de 4 dimensões para 2 componentes principais (`pca1`, `pca2`)
- Utilizado apenas para visualização dos clusters em 2D

### 7. 🎨 Visualização Final

## 📊 Análise dos Resultados

O script imprime uma tabela com os resultados de todas as métricas para K entre 2 e 10, seguida de uma explicação:

```text
📊 Como avaliar os resultados por K:

Cada linha da tabela representa um número de clusters (K) e os valores de desempenho para 5 métricas:

➡️ Silhouette Score: mede o quão bem os pontos estão agrupados. Quanto MAIOR, melhor.
➡️ Calinski-Harabasz Index: mede a separação entre os clusters. Quanto MAIOR, melhor.
➡️ Davies-Bouldin Index: mede a similaridade entre clusters. Quanto MENOR, melhor.
➡️ Intra-cluster Distance: mede a coesão interna. Quanto MENOR, melhor.
➡️ Inter-cluster Distance: mede a separação entre clusters. Quanto MAIOR, melhor.

✅ K = 2 se destaca como o melhor em praticamente todas as métricas:
   - Maior Silhouette (0.58)
   - Maior Calinski (251)
   - Menor Davies-Bouldin (0.59)
   - Maior separação entre clusters (3.36)

⚠️ Porém, o dataset Iris tem 3 espécies reais. Se você quer refletir essa estrutura real, K = 3 também é uma boa escolha, mesmo que tenha desempenho um pouco inferior nas métricas.

🎯 Em resumo:
- Use K = 2 para máxima performance matemática.
- Use K = 3 para refletir a separação real das espécies do Iris.


````python
# 🔁 Projeto completo: Iris + KMeans + PCA + Múltiplas Métricas + Explicações Gráficas

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import (
    silhouette_score,
    calinski_harabasz_score,
    davies_bouldin_score
)
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
os.environ["LOKY_MAX_CPU_COUNT"] = "4"

# 📥 Carregar o dataset Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# 🔍 Análise exploratória
print("📌 .info():")
print(df.info())
print("\n📌 Missing values:")
print(df.isnull().sum())
print("\n📌 Estatísticas descritivas:")
print(df.describe())
print("\n📌 Correlação:")
print(df.corr())
print("\n📌 Primeiras linhas:")
print(df.head())

# 📈 Visualizações iniciais
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='viridis')
plt.title("Correlação entre variáveis - Iris")
plt.show()

sns.pairplot(df)
plt.suptitle("Dispersões e distribuições - Dataset Iris", y=1.02)
plt.show()

# ⚙️ Normalização
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# 📉 Método do Cotovelo (WCSS)
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Número de Clusters (K)")
plt.ylabel("WCSS")
plt.title("Método do Cotovelo - Iris")
plt.grid(True)
plt.show()

print("""
📉 Método do Cotovelo (Elbow Method):
Esse gráfico mostra a soma das distâncias quadradas dentro dos clusters (WCSS) para diferentes valores de K.

✅ O ponto ideal é onde ocorre uma "curva" ou "cotovelo" — ou seja, onde o ganho de adicionar novos clusters passa a ser marginal.
❌ Após esse ponto, continuar aumentando o número de clusters reduz muito pouco o WCSS e pode gerar overfitting.

🔍 Use o gráfico para visualizar onde o WCSS começa a diminuir mais lentamente.
""")

# 📊 Avaliação de múltiplas métricas
k_values = range(2, 11)
silhouette_scores = []
calinski_scores = []
davies_scores = []
intra_cluster_distances = []
inter_cluster_distances = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(scaled_data)

    silhouette_scores.append(silhouette_score(scaled_data, labels))
    calinski_scores.append(calinski_harabasz_score(scaled_data, labels))
    davies_scores.append(davies_bouldin_score(scaled_data, labels))

    intra_dist = np.mean([
        np.linalg.norm(scaled_data[i] - kmeans.cluster_centers_[label])
        for i, label in enumerate(labels)
    ])
    intra_cluster_distances.append(intra_dist)

    centers = kmeans.cluster_centers_
    inter_distances = [
        np.linalg.norm(a - b)
        for i, a in enumerate(centers)
        for j, b in enumerate(centers) if i < j
    ]
    inter_cluster_distances.append(np.mean(inter_distances))

# 🎯 Gráfico + explicação: Silhouette Score
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores, marker='o', color='green')
plt.title("Silhouette Score")
plt.xlabel("Número de Clusters (K)")
plt.ylabel("Score")
plt.grid(True)
plt.show()
print("""
🟢 Silhouette Score:
Esse gráfico mostra o quão bem os pontos estão agrupados.

✅ Quanto mais próximo de 1, melhor o agrupamento.
❌ Valores abaixo de 0.5 indicam que os clusters podem estar se sobrepondo.
🔍 Escolha o valor de K com o maior Silhouette Score.
""")

# 🎯 Gráfico + explicação: Calinski-Harabasz
plt.figure(figsize=(8, 5))
plt.plot(k_values, calinski_scores, marker='o', color='blue')
plt.title("Calinski-Harabasz Index")
plt.xlabel("Número de Clusters (K)")
plt.ylabel("Score")
plt.grid(True)
plt.show()
print("""
🔵 Calinski-Harabasz Index:
Avalia a separação entre clusters e a compacidade dentro de cada grupo.

✅ Quanto maior o valor, melhor os clusters estão separados e definidos.
📌 Ótimo para reforçar a escolha de K que aparece em outras métricas.
""")

# 🎯 Gráfico + explicação: Davies-Bouldin
plt.figure(figsize=(8, 5))
plt.plot(k_values, davies_scores, marker='o', color='red')
plt.title("Davies-Bouldin Index")
plt.xlabel("Número de Clusters (K)")
plt.ylabel("Índice")
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
print("""
🔴 Davies-Bouldin Index:
Mede a similaridade entre os clusters.

✅ Quanto menor o índice, melhor.
❌ Valores altos significam que os clusters estão muito próximos e mal definidos.
🔍 Escolha o menor valor possível para definir o melhor K.
""")

# 🎯 Gráfico + explicação: Distância Intra-Cluster
plt.figure(figsize=(8, 5))
plt.plot(k_values, intra_cluster_distances, marker='o', color='purple')
plt.title("Distância Intra-Cluster Média")
plt.xlabel("Número de Clusters (K)")
plt.ylabel("Distância Média")
plt.grid(True)
plt.show()
print("""
🟣 Distância Intra-Cluster:
Mede o quão próximos os pontos estão do centro de seu cluster.

✅ Valores baixos indicam que os pontos estão bem agrupados (alta coesão).
❌ Reduções muito pequenas ao aumentar K indicam que mais clusters não estão ajudando.
""")

# 🎯 Gráfico + explicação: Distância Inter-Cluster
plt.figure(figsize=(8, 5))
plt.plot(k_values, inter_cluster_distances, marker='o', color='orange')
plt.title("Distância Inter-Cluster Média")
plt.xlabel("Número de Clusters (K)")
plt.ylabel("Distância Média")
plt.grid(True)
plt.show()
print("""
🟠 Distância Inter-Cluster:
Mede o quão distantes os clusters estão entre si.

✅ Quanto maior a distância, mais separados estão os grupos.
❌ Valores baixos sugerem sobreposição entre clusters.
🔍 Boa para avaliar se os clusters são bem distintos.
""")

# 📋 Resultados numéricos
resultados_df = pd.DataFrame({
    "K": list(k_values),
    "Silhouette": silhouette_scores,
    "Calinski-Harabasz": calinski_scores,
    "Davies-Bouldin": davies_scores,
    "Intra-cluster": intra_cluster_distances,
    "Inter-cluster": inter_cluster_distances
})
print("\n📊 Tabela de Resultados por K:")
print(resultados_df)

print("""
📊 Como avaliar os resultados por K:

Cada linha da tabela representa um número de clusters (K) e os valores de desempenho para 5 métricas:

➡️ Silhouette Score: mede o quão bem os pontos estão agrupados. Quanto MAIOR, melhor.
➡️ Calinski-Harabasz Index: mede a separação entre os clusters. Quanto MAIOR, melhor.
➡️ Davies-Bouldin Index: mede a similaridade entre clusters. Quanto MENOR, melhor.
➡️ Intra-cluster Distance: mede a coesão interna. Quanto MENOR, melhor.
➡️ Inter-cluster Distance: mede a separação entre clusters. Quanto MAIOR, melhor.

✅ K = 2 se destaca como o melhor em praticamente todas as métricas:
   - Maior Silhouette (0.58)
   - Maior Calinski (251)
   - Menor Davies-Bouldin (0.59)
   - Maior distância inter-cluster (3.36)

⚠️ Porém, o dataset Iris tem 3 espécies reais. Se você quer refletir essa estrutura real, K = 3 também é uma boa escolha, mesmo que tenha desempenho um pouco inferior nas métricas.

🎯 Em resumo:
- Use K = 2 se quiser clusters matematicamente mais coesos e bem separados.
- Use K = 3 se quiser refletir a separação real das espécies do Iris.
""")

# ✅ Melhor K com base no Silhouette
best_k = resultados_df["Silhouette"].idxmax() + 2
print(f"\n🔥 Melhor K com base no Silhouette Score: {best_k}")

# 🧠 Aplicar KMeans final
final_kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df['cluster'] = final_kmeans.fit_predict(scaled_data)

# 🔄 PCA para reduzir de 4D para 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
df['pca1'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]

# 🎨 Visualização dos clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='pca1', y='pca2', hue='cluster', palette='tab10')
plt.title(f"Clusters com PCA - Iris (K = {best_k})")
plt.show()
````
