# Gerador de Imagens com Redes Neurais Adversariais (GAN)

## ğŸ“Œ DescriÃ§Ã£o
Este projeto implementa uma Rede Neural Adversarial Generativa (GAN) para a geraÃ§Ã£o de imagens sintÃ©ticas. As GANs sÃ£o compostas por dois modelos concorrentes: um gerador e um discriminador. O gerador cria imagens sintÃ©ticas, enquanto o discriminador tenta distinguir entre imagens reais e sintÃ©ticas. O treinamento ocorre de forma adversarial, resultando em imagens cada vez mais realistas.

## ğŸ“Œ Estrutura do CÃ³digo
O cÃ³digo Ã© estruturado da seguinte forma:

1. **ImportaÃ§Ã£o de bibliotecas** - Utiliza o PyTorch para construir e treinar a GAN.
2. **DefiniÃ§Ã£o do gerador e discriminador** - Modelos de redes neurais que compÃµem a GAN.
3. **FunÃ§Ãµes de treinamento** - ImplementaÃ§Ã£o do loop de treinamento para aperfeiÃ§oar os modelos.
4. **GeraÃ§Ã£o de imagens** - Uso do gerador treinado para criar novas imagens.

````python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# ğŸ”¹ ConfiguraÃ§Ãµes
BATCH_SIZE = 128
EPOCHS = 100
LATENT_DIM = 100  # Tamanho do vetor de entrada do gerador
IMG_SIZE = 28
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ğŸ”¹ Baixar e preparar o dataset MNIST
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_loader = torch.utils.data.DataLoader(
    torchvision.datasets.MNIST("./data", train=True, download=True, transform=transform),
    batch_size=BATCH_SIZE, shuffle=True
)

# ğŸ”¹ Criando o Gerador (Generator)
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(LATENT_DIM, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, IMG_SIZE * IMG_SIZE),
            nn.Tanh()
        )
    
    def forward(self, z):
        return self.model(z).view(-1, 1, IMG_SIZE, IMG_SIZE)

# ğŸ”¹ Criando o Discriminador (Discriminator)
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(IMG_SIZE * IMG_SIZE, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )
    
    def forward(self, img):
        return self.model(img.view(-1, IMG_SIZE * IMG_SIZE))

# ğŸ”¹ Inicializar modelos, otimizadores e funÃ§Ã£o de perda
G = Generator().to(DEVICE)
D = Discriminator().to(DEVICE)
criterion = nn.BCELoss()
optimizer_G = optim.Adam(G.parameters(), lr=0.0002)
optimizer_D = optim.Adam(D.parameters(), lr=0.0002)

# ğŸ”¹ Treinamento da GAN
for epoch in range(EPOCHS):
    for i, (imgs, _) in enumerate(train_loader):
        real_imgs = imgs.to(DEVICE)
        batch_size = real_imgs.shape[0]

        # Criando rÃ³tulos reais e falsos
        real_labels = torch.ones(batch_size, 1).to(DEVICE)
        fake_labels = torch.zeros(batch_size, 1).to(DEVICE)

        # ğŸ”¹ Treinar o Discriminador
        optimizer_D.zero_grad()
        real_loss = criterion(D(real_imgs), real_labels)
        z = torch.randn(batch_size, LATENT_DIM).to(DEVICE)
        fake_imgs = G(z)
        fake_loss = criterion(D(fake_imgs.detach()), fake_labels)
        loss_D = real_loss + fake_loss
        loss_D.backward()
        optimizer_D.step()

        # ğŸ”¹ Treinar o Gerador
        optimizer_G.zero_grad()
        fake_loss_G = criterion(D(fake_imgs), real_labels)
        fake_loss_G.backward()
        optimizer_G.step()

    # ğŸ”¹ Exibir progresso do treinamento
    print(f"Epoch [{epoch+1}/{EPOCHS}] | Loss_D: {loss_D.item():.4f} | Loss_G: {fake_loss_G.item():.4f}")
    if (epoch + 1) % 10 == 0:
        with torch.no_grad():
            z = torch.randn(16, LATENT_DIM).to(DEVICE)
            samples = G(z).cpu()
            fig, axes = plt.subplots(1, 16, figsize=(16, 2))
            for j in range(16):
                axes[j].imshow(samples[j].squeeze(), cmap='gray')
                axes[j].axis("off")
            plt.show()
````

# Gerador de Imagens com Stable Diffusion

Este projeto utiliza o modelo **Stable Diffusion v2** para gerar imagens a partir de descriÃ§Ãµes textuais. O modelo Ã© carregado a partir do repositÃ³rio da **Hugging Face** e utiliza **PyTorch** para inferÃªncia otimizada.

## ğŸš€ Como Usar

1. **Instale as dependÃªncias necessÃ¡rias:**
    ```sh
    pip install torch diffusers transformers matplotlib
    ```
2. **Execute o script:**
    ```sh
    python stable_diffusion_generator.py
    ```
3. **Digite a descriÃ§Ã£o da imagem desejada** quando solicitado.
4. **Aguarde a geraÃ§Ã£o da imagem** (processo pode levar alguns segundos).
5. **A imagem serÃ¡ exibida automaticamente e salva como** `imagem_gerada.png`.

## ğŸ“Œ ParÃ¢metros AjustÃ¡veis

- `guidance_scale`: Ajusta a fidelidade da imagem ao prompt (recomendado entre 8.0 e 12.0).
- `num_inference_steps`: NÃºmero de passos de inferÃªncia para melhor qualidade (padrÃ£o 100).
- `width` e `height`: Definem a resoluÃ§Ã£o da imagem (768x768 por padrÃ£o, pode ser aumentado para 1024x1024).

## ğŸ›  Requisitos

- Python 3.8+
- GPU compatÃ­vel com CUDA para melhor desempenho

## ğŸ“· Exemplo de Uso

Entrada:
```sh
Digite uma descriÃ§Ã£o para a imagem: "Um dragÃ£o robÃ³tico em uma cidade cyberpunk"
```

SaÃ­da:
- Uma imagem gerada baseada na descriÃ§Ã£o fornecida serÃ¡ salva como `imagem_gerada.png`.

## ğŸï¸ OtimizaÃ§Ãµes para Melhor Performance

- **LiberaÃ§Ã£o de MemÃ³ria:** O cÃ³digo libera VRAM antes e depois da geraÃ§Ã£o para evitar lentidÃ£o.
- **Uso de Float16:** Reduz o uso de VRAM, permitindo um desempenho mais eficiente.
- **OtimizaÃ§Ã£o de CUDA:** `torch.backends.cudnn.benchmark = True` melhora a velocidade de inferÃªncia.

## ğŸ“– ReferÃªncias
- [Hugging Face Diffusers](https://huggingface.co/docs/diffusers)
- [Stable Diffusion Model](https://huggingface.co/stabilityai/stable-diffusion-2)

## ğŸ“œ LicenÃ§a
Este projeto segue a licenÃ§a do modelo **Stable Diffusion** e da biblioteca **diffusers** da Hugging Face.



````python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# OtimizaÃ§Ã£o do CUDA para melhor desempenho
torch.backends.cudnn.benchmark = True

# Verifica se hÃ¡ GPU disponÃ­vel para acelerar a geraÃ§Ã£o
device = "cuda" if torch.cuda.is_available() else "cpu"

# Carregar o modelo com otimizaÃ§Ã£o para melhor qualidade
modelo_nome = "stabilityai/stable-diffusion-2"
pipeline = StableDiffusionPipeline.from_pretrained(
    modelo_nome,
    torch_dtype=torch.float16,  # Usa menos VRAM para desempenho melhor
    revision="fp16"
).to(device)

def liberar_memoria():
    """Libera VRAM antes e depois da geraÃ§Ã£o para evitar lentidÃ£o e travamentos."""
    torch.cuda.empty_cache()
    torch.cuda.ipc_collect()

def gerar_imagem(prompt, guidance_scale=10.0, num_inference_steps=100, width=768, height=768):
    """
    Gera uma imagem a partir de um prompt de texto usando Stable Diffusion v2.

    ParÃ¢metros:
    - prompt (str): DescriÃ§Ã£o da imagem desejada.
    - guidance_scale (float): Ajusta a fidelidade da imagem ao prompt (recomendado entre 8.0 e 12.0).
    - num_inference_steps (int): NÃºmero de passos de inferÃªncia para melhor qualidade (padrÃ£o 100).
    - width (int): Largura da imagem gerada (padrÃ£o 768, pode ser 1024 para mais qualidade).
    - height (int): Altura da imagem gerada (padrÃ£o 768, pode ser 1024 para mais qualidade).
    """
    liberar_memoria()  # Libera VRAM antes da execuÃ§Ã£o

    with torch.no_grad():
        imagem = pipeline(
            prompt, 
            guidance_scale=guidance_scale, 
            num_inference_steps=num_inference_steps, 
            width=width, 
            height=height
        ).images[0]

    # Salvar e exibir a imagem gerada
    imagem.save("imagem_gerada.png")
    print(f"ğŸ“· Imagem gerada e salva como 'imagem_gerada.png' ({width}x{height})")

    liberar_memoria()  # Libera VRAM apÃ³s a execuÃ§Ã£o para evitar travamentos

    plt.imshow(imagem)
    plt.axis("off")
    plt.show()

if __name__ == "__main__":
    while True:
        prompt = input("Digite uma descriÃ§Ã£o para a imagem ('sair' para encerrar): ")
        if prompt.lower() == "sair":
            break
        gerar_imagem(prompt)
````
