# Gerador de Imagens com Redes Neurais Adversariais (GAN)

## 📌 Descrição
Este projeto implementa uma Rede Neural Adversarial Generativa (GAN) para a geração de imagens sintéticas. As GANs são compostas por dois modelos concorrentes: um gerador e um discriminador. O gerador cria imagens sintéticas, enquanto o discriminador tenta distinguir entre imagens reais e sintéticas. O treinamento ocorre de forma adversarial, resultando em imagens cada vez mais realistas.

## 📌 Estrutura do Código
O código é estruturado da seguinte forma:

1. **Importação de bibliotecas** - Utiliza o PyTorch para construir e treinar a GAN.
2. **Definição do gerador e discriminador** - Modelos de redes neurais que compõem a GAN.
3. **Funções de treinamento** - Implementação do loop de treinamento para aperfeiçoar os modelos.
4. **Geração de imagens** - Uso do gerador treinado para criar novas imagens.

````python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 🔹 Configurações
BATCH_SIZE = 128
EPOCHS = 100
LATENT_DIM = 100  # Tamanho do vetor de entrada do gerador
IMG_SIZE = 28
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 🔹 Baixar e preparar o dataset MNIST
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_loader = torch.utils.data.DataLoader(
    torchvision.datasets.MNIST("./data", train=True, download=True, transform=transform),
    batch_size=BATCH_SIZE, shuffle=True
)

# 🔹 Criando o Gerador (Generator)
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(LATENT_DIM, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, IMG_SIZE * IMG_SIZE),
            nn.Tanh()
        )
    
    def forward(self, z):
        return self.model(z).view(-1, 1, IMG_SIZE, IMG_SIZE)

# 🔹 Criando o Discriminador (Discriminator)
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(IMG_SIZE * IMG_SIZE, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )
    
    def forward(self, img):
        return self.model(img.view(-1, IMG_SIZE * IMG_SIZE))

# 🔹 Inicializar modelos, otimizadores e função de perda
G = Generator().to(DEVICE)
D = Discriminator().to(DEVICE)
criterion = nn.BCELoss()
optimizer_G = optim.Adam(G.parameters(), lr=0.0002)
optimizer_D = optim.Adam(D.parameters(), lr=0.0002)

# 🔹 Treinamento da GAN
for epoch in range(EPOCHS):
    for i, (imgs, _) in enumerate(train_loader):
        real_imgs = imgs.to(DEVICE)
        batch_size = real_imgs.shape[0]

        # Criando rótulos reais e falsos
        real_labels = torch.ones(batch_size, 1).to(DEVICE)
        fake_labels = torch.zeros(batch_size, 1).to(DEVICE)

        # 🔹 Treinar o Discriminador
        optimizer_D.zero_grad()
        real_loss = criterion(D(real_imgs), real_labels)
        z = torch.randn(batch_size, LATENT_DIM).to(DEVICE)
        fake_imgs = G(z)
        fake_loss = criterion(D(fake_imgs.detach()), fake_labels)
        loss_D = real_loss + fake_loss
        loss_D.backward()
        optimizer_D.step()

        # 🔹 Treinar o Gerador
        optimizer_G.zero_grad()
        fake_loss_G = criterion(D(fake_imgs), real_labels)
        fake_loss_G.backward()
        optimizer_G.step()

    # 🔹 Exibir progresso do treinamento
    print(f"Epoch [{epoch+1}/{EPOCHS}] | Loss_D: {loss_D.item():.4f} | Loss_G: {fake_loss_G.item():.4f}")
    if (epoch + 1) % 10 == 0:
        with torch.no_grad():
            z = torch.randn(16, LATENT_DIM).to(DEVICE)
            samples = G(z).cpu()
            fig, axes = plt.subplots(1, 16, figsize=(16, 2))
            for j in range(16):
                axes[j].imshow(samples[j].squeeze(), cmap='gray')
                axes[j].axis("off")
            plt.show()
````

# Gerador de Imagens com Stable Diffusion

Este projeto utiliza o modelo **Stable Diffusion v2** para gerar imagens a partir de descrições textuais. O modelo é carregado a partir do repositório da **Hugging Face** e utiliza **PyTorch** para inferência otimizada.

## 🚀 Como Usar

1. **Instale as dependências necessárias:**
    ```sh
    pip install torch diffusers transformers matplotlib
    ```
2. **Execute o script:**
    ```sh
    python stable_diffusion_generator.py
    ```
3. **Digite a descrição da imagem desejada** quando solicitado.
4. **Aguarde a geração da imagem** (processo pode levar alguns segundos).
5. **A imagem será exibida automaticamente e salva como** `imagem_gerada.png`.

## 📌 Parâmetros Ajustáveis

- `guidance_scale`: Ajusta a fidelidade da imagem ao prompt (recomendado entre 8.0 e 12.0).
- `num_inference_steps`: Número de passos de inferência para melhor qualidade (padrão 100).
- `width` e `height`: Definem a resolução da imagem (768x768 por padrão, pode ser aumentado para 1024x1024).

## 🛠 Requisitos

- Python 3.8+
- GPU compatível com CUDA para melhor desempenho

## 📷 Exemplo de Uso

Entrada:
```sh
Digite uma descrição para a imagem: "Um dragão robótico em uma cidade cyberpunk"
```

Saída:
- Uma imagem gerada baseada na descrição fornecida será salva como `imagem_gerada.png`.

## 🏎️ Otimizações para Melhor Performance

- **Liberação de Memória:** O código libera VRAM antes e depois da geração para evitar lentidão.
- **Uso de Float16:** Reduz o uso de VRAM, permitindo um desempenho mais eficiente.
- **Otimização de CUDA:** `torch.backends.cudnn.benchmark = True` melhora a velocidade de inferência.

## 📖 Referências
- [Hugging Face Diffusers](https://huggingface.co/docs/diffusers)
- [Stable Diffusion Model](https://huggingface.co/stabilityai/stable-diffusion-2)

## 📜 Licença
Este projeto segue a licença do modelo **Stable Diffusion** e da biblioteca **diffusers** da Hugging Face.



````python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import os
import warnings
from tqdm import TqdmWarning
os.environ["IPYTHONENABLE"] = "0"
warnings.filterwarnings("ignore", category=TqdmWarning)

# Otimização do CUDA para melhor desempenho
torch.backends.cudnn.benchmark = True

# Verifica se há GPU disponível para acelerar a geração
device = "cuda" if torch.cuda.is_available() else "cpu"

# Carregar o modelo com otimização para melhor qualidade
modelo_nome = "stabilityai/stable-diffusion-2"
pipeline = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1",
    torch_dtype=torch.float16,
    use_safetensors=True
).to(device)

pipeline.enable_attention_slicing()


def liberar_memoria():
    """Libera VRAM antes e depois da geração para evitar lentidão e travamentos."""
    torch.cuda.empty_cache()
    torch.cuda.ipc_collect()

def gerar_imagem(prompt, guidance_scale=10.0, num_inference_steps=100, width=768, height=768):
    """
    Gera uma imagem a partir de um prompt de texto usando Stable Diffusion v2.

    Parâmetros:
    - prompt (str): Descrição da imagem desejada.
    - guidance_scale (float): Ajusta a fidelidade da imagem ao prompt (recomendado entre 8.0 e 12.0).
    - num_inference_steps (int): Número de passos de inferência para melhor qualidade (padrão 100).
    - width (int): Largura da imagem gerada (padrão 768, pode ser 1024 para mais qualidade).
    - height (int): Altura da imagem gerada (padrão 768, pode ser 1024 para mais qualidade).
    """
    liberar_memoria()  # Libera VRAM antes da execução

    with torch.no_grad():
        imagem = pipeline(
            prompt, 
            guidance_scale=guidance_scale, 
            num_inference_steps=num_inference_steps, 
            width=width, 
            height=height
        ).images[0]

    # Salvar e exibir a imagem gerada
    imagem.save("imagem_gerada.png")
    print(f"📷 Imagem gerada e salva como 'imagem_gerada.png' ({width}x{height})")

    liberar_memoria()  # Libera VRAM após a execução para evitar travamentos

    plt.imshow(imagem)
    plt.axis("off")
    plt.show()

if __name__ == "__main__":
    while True:
        prompt = input("Digite uma descrição para a imagem ('sair' para encerrar): ")
        if prompt.lower() == "sair":
            break
        gerar_imagem(prompt)
````
